# Descriptive Statistics - Mean, Mode, Median, standard Deviation, Variance and Inter Quartile Range

## Measurment of Central Tendency

### Mean

```python
student_marks = [100,82,90,18,25,28,35,45,51,62,45]
```

```python
avg_marks = sts.mean(student_marks)
avg_marks
```

### Median

```python
avg_marks = sts.median(student_marks)
avg_marks
```

### Mode

```python
fquency = sts.mode(student_marks)
fquency
```

## Measurment of Dispersion

### Standard Deviation

```python
stdeviation_marks = sts.stdev(student_marks)
stdeviation_marks
```

### Variance

```python
var_marks = sts.variance(student_marks)
var_marks
```

### Inter Quartile Range IQR

```python
quantiles = sts.quantiles(student_marks)
quantiles
```

```python
q1 = quantiles[0]
q2 = quantiles[1]
q3 = quantiles[2]
print("0-25th percentile",q1)
print("25-50 percentile",q2)
print("50-75 percentile",q3)
print()
print("""Q1 – Lower Quartile Part
Q2 – Median
Q3 – Upper Quartile Part""")
```

```python
# Interquartile range = Upper Quartile – Lower Quartile
IQR = q3-q1
print("Interquartile range",IQR)
```

probability of getting 2 defects out of 20 products


where
- k = 2 defects we are looking (range=0-2)
- n = 20 total items to be checked
- p = 0.12 


```python
binom.cdf(2,20,0.12)
```

```python
results = binom.cdf(2,20,0.12)
print("Probability of getting 2 or less than 2 defects in a sample of 20: ",results)
```

when we need to calculate probability of getting exactly 2 defects out of 20 we use binom.pmf()

```python
results = binom.pmf(2,20,0.12)
print("Probability of getting exactly 2 defects in a sample of 20: ",results)
```

probability of getting more than 2 defects

```python
results = 1-binom.cdf(2,20,0.12)
print("Probability of getting more than 2 defects in a sample of 20 :",results)
```

Or

```python
results = binom.sf(2,20,0.12)
print("Probability of getting more than 2 defects in a sample of 20 :",results)
```

On a booking counter on the average 3.6 
people come every 10 minute on weekends. 
What is the probability of getting 7 people in 
10 minutes?

```python
p_7_people = poisson.pmf(7,3.6)
print("Probability of getting exactly 7 people is :",p_7_people)
p_7_or_more_people = poisson.sf(7,3.6)
print("Probability of getting more than 7 people is :",p_7_or_more_people)
p_7_or_less_people = poisson.cdf(7,3.6)
print("Probability of getting 0 to  7 (less than 7) people is :",p_7_or_less_people)
```

```python
expectation = poisson.mean(3.6)
var = poisson.var(3.6)
std = poisson.std(3.6)
print("The expectation or average of people coming next weekend is: ",expectation)
print("The variance of people coming next weekend is: ",var)
print("The standard deviation of people coming next weekend is: ",std)
```

### Using SciPy

 Perfume bottles are filled with the 
average volume of 150 cc and the 
standard deviation of 2 cc.


-  What percent of bottles will have 
the volume more than 153 cc?


```python
# norm.sf(x,mu,std)
```

```python
p_more_than_153cc = norm.sf(153,150,2)
print("Probability of bottles with volume more than 153cc is: ",p_more_than_153cc)
print("Percentage of bottles with volume more than 153cc is: ",(p_more_than_153cc*100))
```

What percent of bottles will have 
the volume between 148 and 152 
cc

```python
# We know it is 68% bcz the AUC falls btw +1 and -1
```

```python
p_btw_148_and_152cc = norm.sf(148,150,2) - norm.sf(152,150,2)
print("Percentage of bottles with between 148cc and 152cc is:",(p_btw_148_and_152cc*100))
```

pdf is similar to pmf but it is used in Normal Distriibution

```python
pdf = norm.pdf(148,150,2)
print("The probability of getting exactly the volume of 148cc: ",pdf)
more_than_152cc = norm.sf(152,150,2)
print("The probability of getting more than 152cc volume is: ",more_than_152cc)
```

```python
# norm.sf(x,mu,std) - for more than (1 - cdf)
# norm.cdf(x,mu,std) - for less than
# norm.pdf(x,mu,std) - 
# norm.mean(mu)
# norm.var(mu)
# norm.std(mu)
```

## Types of t Test

If two set of data are independent or dependent.
- If the values in one sample reveal no information about those of the other sample, then the samples are independent.
- Example:  Volume produced by two machines
- **Use Two sample t test**
- If the values in one sample affect the values in the other sample, then the samples are dependent. 
- Example: Blood pressure before and after a specific medicine
- **Use Paired t test**

- 1 Equal Variance two sample
- 2 UnEqual Variance two sample

- Ha: μA ≠ μB

```python
from scipy.stats import f_oneway
f_cal,pvalue = f_oneway(machine_A,machine_B,machine_C)
print(f_cal)
print(pvalue)
if pvalue > 0.05:
    print("Failed to reject Null Hypothesis")
else:
    print("Reject Null Hypothesis")
```

```python
from statsmodels.stats import oneway
```

```python
oneway.anova_oneway((machine_A,machine_B,machine_C),use_var='equal')
```

## Goodness of Fit Test (chisquare)

```python
expected_coin_trails = [50,50] #heads,tails
observed_coin_trails = [40,60]
statistic,pvalue=stats.chisquare(observed_coin_trails,expected_coin_trails)
print(statistic)
print(pvalue)
if pvalue > 0.05:
    print("Failde to reject null hypothesis because here coin is not biased")
else:
    print("Reject null hypothesis because here coin is biased")
```

```python
data = sns.load_dataset("tips")
data
```

- Null Hypothesis: There is no relation between day and smoker

- Alternate Hypothesis: There is  relation between day and smoker

```python
data.groupby(["day","smoker"]).count()
```

```python
day_smoke = data.pivot_table(index='day',columns='smoker',aggfunc='count')['tip']
day_smoke
```

```python
stats.chi2_contingency(day_smoke)
```

**There is  relation between day and smoker**
